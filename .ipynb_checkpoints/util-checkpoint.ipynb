# get review data from combined title

```python
def loop_get_review(title_list):
    first_iter = True
    for title in title_list:
        try:
            if first_iter:
                df_review = get_movie_reviews(title)
                first_iter = False
            else:
                df_review_ = get_movie_reviews(title)
                df_review = pd.concat([df_review,df_review_], ignore_index = True) 
        except Exception as e:
            print(f'Failed to get review from {title}: {e}')
    # write df to csv
    first_title = title_list[0]
    last_title = title_list[1]
    df_review.to_csv(f'review_data_{first_title}_{last_title}.csv', index = False)
    return df_review
```
